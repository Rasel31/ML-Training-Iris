```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

# Iris Dataset

```python
iris
```

```python
correlation = iris.corr()
```

```python
correlation
```

# COLOR PLOT

```python
sns.heatmap(iris.corr(), cmap = 'coolwarm', annot = True)
```

# BEST FIT WITH LINE

```python
sns.set(style = "whitegrid", color_codes = True) # white, dark, whitegrid, darkgrid, ticks
```

```python
sns.lmplot(x = "sepal_length", y = "petal_length", hue = "species", data = iris, \
          palette = "Set1", fit_reg = True, markers = ["o", "s", "d"])
```

# BEST FIT WITHOUT LINE

```python
sns.lmplot(x = "sepal_length", y = "petal_length", hue = "species", data = iris, \
          palette = "Set1", fit_reg = False, markers = ["o", "s", "d"])
```

```python
sns.scatterplot(x = "sepal_length", y= "sepal_width", hue = "species", style = "species", data = iris)
```

```python
sns.lmplot(x="sepal_length", y="sepal_width", data=iris, hue="species", legend=True)
```

```python
sns.lmplot(x="sepal_length", y="sepal_width", data=iris, hue="species", legend=False)
```

# PAIR PLOT OF TWO FEATURES

```python
sns.pairplot(iris, height = 5, vars = ["sepal_length", "petal_length"], markers = ["o", "s", "d"], hue = "species")
```

# PAIR PLOT OF ALL FEATURES

```python
sns.pairplot(iris, markers=["o", "s", "D"], hue="species")
```

```python
sns.pairplot(iris, kind="reg", hue="species", markers=["o", "s", "D"])
```

# BOX PLOT

```python
del iris['ID']
del iris['ratio']
```

```python
sns.boxplot(data = iris, orient = "h")
```

```python
iris.index
```

```python
iris["ID"] = iris.index
iris["ratio"] = iris["sepal_length"]/iris["sepal_width"]
sns.lmplot(x="ID", y="ratio", data=iris, hue="species", markers=["o", "s", "D"], fit_reg=False, legend=True)
```

# STRIP PLOT

```python
sns.stripplot(x = "species", y = "sepal_length", data = iris, jitter = True)
```

```python
sns.stripplot(x = "species", y = "petal_length", data = iris, jitter = False)
```

# SWARM PLOT

```python
sns.swarmplot(x="species", y="petal_length", data=iris)
```
